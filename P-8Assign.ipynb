```python

class CustomError(Exception):
    def __init__(self, message):
        super().__init__(message)


def check_value(value):
    if value < 0:
        raise CustomError("Negative values are not allowed!")
    elif value == 0:
        raise CustomError("Zero is not a valid input!")
    else:
        print(f"Valid input: {value}")


try:
    num = int(input("Enter a number: "))
    check_value(num)
except CustomError as e:
    print(f"Custom Exception Caught: {e}")
except ValueError:
    print("Invalid input! Please enter a valid number.")
```

```python

elements = [10, 20, 30, 40, 50]


def access_element():
    while True:
        try:
            index = int(input("Enter an index to access (0-4): "))
            print(f"Element at index {index}: {elements[index]}")
            break
        except IndexError:
            print("Error: Index out of range! Please enter a valid index between 0 and 4.")
        except ValueError:
            print("Error: Invalid input! Please enter a numerical index.")


access_element()
```

```python

def convert_to_integer():
    while True:
        try:
            user_input = input("Enter a number: ")
            number = int(user_input)
            print(f"Valid integer entered: {number}")
            break
        except ValueError:
            print("Error: Invalid input! Please enter a valid integer.")

convert_to_integer()
```
